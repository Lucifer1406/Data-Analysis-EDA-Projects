# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset

```python
df = pd.read_csv('Netflix.csv')
```

```python
df.head(5)
```

# Data Wrangling and Data Manipulation 

```python
# Data Information
df.info()
```

```python
# Chech the missing values from given data
df.isnull().sum()
```

### Drop Duplicate values from data

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

### Handling Missing Values of the given data

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['date_added'] = df['date_added'].ffill()  # forward fill method
df['rating'] = df['rating'].fillna('Unknown')
```

```python
df.head(5)
```

```python
df.tail(5)
```

```python
df.info()
```

### Convert date_added to datetime

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')
```

```python
df.head(5)
```

### Clean Duration FIeld

```python
df['duration_num'] = df['duration'].str.extract(r'(\d+)').astype(float)
df['duration_type'] = df['duration'].str.extract(r'([a-zA-Z]+)').fillna('Unknown')
```

```python
df.info()
```

```python
df.head()
```

```python
df.shape
```

# Data Visualization with EDA

```python
sns.set(style='whitegrid')
plt.figure(figsize=(12,6))
```

### Count of Movies VS TV Shows

```python
sns.countplot(data=df, x='type')
plt.title("Count of Movies Vs TV Shows")
plt.show()
```

### Top 10 Countries with most content

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='barh', color='skyblue')
plt.title('Top 10 Countries by Number of Titles')
plt.xlabel('Number of Titles')
plt.gca().invert_yaxis()
plt.show()
```

### Content Added Over Time

```python
df['year_added'] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind='bar', color='salmon')
plt.title('Content Added Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.show()
```

### Distribution of Release Years

```python
plt.figure(figsize=(14, 6))
sns.histplot(df['release_year'], bins=30, kde=False, color='olive')
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

### Rating Distribution

```python
plt.figure(figsize=(10, 6))
df['rating'].value_counts().head(10).plot(kind='bar', color='coral')
plt.title('Top 10 Content Ratings')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### Duration Distribution By Type

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='type', y='duration_num', data=df)
plt.title('Duration Distribution by Type')
plt.show()
```

### Top 10 Genres (Listed In)

```python
from collections import Counter
genres = df['listed_in'].str.split(', ')
flat_genres = [genre for sublist in genres for genre in sublist]
genre_counts = pd.Series(flat_genres).value_counts().head(10)
genre_counts.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Genres')
plt.xticks(rotation=45)
plt.ylabel('Count')
plt.show()
```

### Top 10 Directors 

```python
top_directors = df[df['director'] != 'Unknown']['director'].value_counts().head(10)
top_directors.plot(kind='barh', color='orchid')
plt.title('Top 10 Directors by Content')
plt.xlabel('Number of Titles')
plt.gca().invert_yaxis()
plt.show()
```

### TV Shows Vs Movie Over time

```python
df.groupby(['year_added', 'type']).size().unstack().plot(kind='line', marker='o')
plt.title('TV Shows vs Movies Added Over Years')
plt.ylabel('Number of Titles')
plt.show()
```

### Content Length Distribution (Movies Only)

```python
movies = df[df['type'] == 'Movie']
plt.hist(movies['duration_num'].dropna(), bins=20, color='deepskyblue')
plt.title('Movie Length Distribution (in minutes)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```
